```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns 

import plotly.express as px
import plotly.graph_objects as go
from jupyter_dash import JupyterDash
import dash_core_components as dcc
import dash_html_components as html
from dash.dependencies import Input, Output
```

```python
df= pd.read_csv(r'./Data_Professional_Salary_Survey_Responses.csv')
df.rename(columns={' SalaryUSD ': 'SalaryUSD'}, inplace=True)
#show the first 5 rows
df.head()
```

## Determine how many missing values are present in each column

```python
# replacing missing values with np.nan
missing_val = ['Not Asked']
df.replace(missing_val, np.nan, inplace = True)
```

```python
# Look at total NaN values per column
df.isnull().sum()
```

```python
df.isnull().sum().sum()
```

## Remove columns that contains NaN more than 30%

```python
df=df.drop(['PostalCode', 'HowManyCompanies', 'CompanyEmployeesOverall', 'Education', 'EducationIsComputerRelated', 
'Certifications', 'HoursWorkedPerWeek', 'TelecommuteDaysPerWeek', 'NewestVersionInProduction', 'OldestVersionInProduction',
'PopulationOfLargestCityWithin20Miles', 'OtherJobDuties', 'KindsOfTasksPerformed', 'LookingForAnotherJob'], axis = 1)
```

```python
df.isnull().sum()
```

## Fill nulls values

```python
df['OtherDatabases'] = df['OtherDatabases'].fillna(df['OtherDatabases'].mode()[0])
df['DatabaseServers'] = df['DatabaseServers'].fillna(df['DatabaseServers'].mode()[0])
df['CareerPlansThisYear'] = df['CareerPlansThisYear'].fillna(df['CareerPlansThisYear'].mode()[0])
# We can't fill the Gender by mode. We assume that it is Unknown
df['Gender'] = df['Gender'].fillna('Unknown')
df['Gender']= df['Gender'].replace(['None'],'Unknown')
```

```python
# Build App
app = JupyterDash(__name__)

years= [2021, 2020, 2019, 2018, 2017]
country = ['Sweden', 'United States', 'United Kingdom', 'Pakistan',
       'Paraguay', 'India', 'Serbia and Montenegro', 'Germany', 'Canada',
       'Dominican Republic', 'Slovakia', 'Brazil', 'France', 'Poland',
       'Romania', 'Ireland', 'South Africa', 'Portugal', 'Costa Rica',
       'Israel', 'Argentina', 'Peru', 'Australia', 'Venezuela', 'Spain',
       'Belgium', 'Malaysia', 'Netherlands', 'Russia', 'Italy', 'Finland',
       'Turkey', 'Austria', 'New Zealand', 'Honduras',
       'United Arab Emirates', 'Guernsey', 'Norway', 'Luxembourg',
       'Puerto Rico', 'Guatemala', 'Greece', 'Denmark', 'Lithuania',
       'Ukraine', 'Philippines', 'Hungary', 'Estonia', 'Cayman Islands',
       'Croatia', 'Indonesia', 'Thailand', 'Switzerland', 'Slovenia',
       'Bosnia and Herzegovina', 'Cyprus', 'Bulgaria', 'Singapore',
       'Colombia', 'Chile', 'Iceland', 'Macedonia', 'Jersey', 'Latvia',
       'Czech Republic', 'Mexico', 'Bangladesh', 'China', 'Iran', 'Qatar',
       'Ecuador', 'Uruguay', 'Saudi Arabia', 'Vanuatu', 'Taiwan',
       'Maldives', 'Armenia', 'Swaziland', 'Georgia', 'Hong Kong',
       'Syria', 'Bahrain', 'Moldova', 'Nicaragua', 'Jamaica', 'Malta',
       'Jordan', 'Uganda', 'Belarus', 'Nepal', 'Ghana', 'El Salvador',
       'Albania', 'Bolivia', 'Vietnam', 'Bermuda', 'Kenya']
fig = px.scatter(df,x="PrimaryDatabase",y="SalaryUSD",color="Country")#IT Draws the Scatter Plot
app.layout = html.Div([
                       
                        html.H2("Most Used Databases"),
                        html.Label([
                                   "Year",
                                   dcc.Dropdown(
                                        id='yrs', 

                                        value= 2021, #it is the default value
                                         options=[{'label':x, "value":x} for x in years],
                                       style={'width': '30%'}#It decreases the drop down menu width
                                        )
                        ]),
                        html.Label([
                                   "Primary Database",
                                   dcc.Dropdown(
                                        id='c', 
                                        clearable=False,
                                        value= "United States", #it is the defaullt value 
                                         options=[{'label':x, "value":x} for x in country],
                                       style={'width': '30%'}#it decreases the dropdown menu width
                                        )
                                  ]),
                        dcc.Graph(id="graph")
                        
                        ,html.Div([html.H2("Title"),dcc.Graph(id="graph2",figure=fig)])#this is 2nd Visualization
])
@app.callback(
    Output('graph', 'figure'),
    [Input("yrs", "value"),
    Input("c","value")]
    
)
def update_figure(val,c):

    test = df[(df['Survey Year']==val) & (df['Country']==c)]
    value_counts = test['PrimaryDatabase'].value_counts()#value counts for primary database
    return px.bar(x=value_counts.index, y=value_counts,labels={"x":" Primary Databases","y":"No of Used Database"})#bar graph
# Note: when we dont have mode="inline" we can open it is a browser
app.run_server( debug=True)
```
